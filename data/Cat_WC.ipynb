```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from PIL import Image
import numpy as np
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
```

```python
cat = pd.read_csv("Caterpillar.csv")
cat
```

```python
cand_mask=np.array(Image.open('캐터필러 유사누끼.png'))
```

```python
ctwc = "".join(cat["Title"].values)
```

```python
plt.figure(figsize=(20,20))
wc = WordCloud(width=1000, height=600, background_color="white",mask=cand_mask)
wc.generate_from_text(ctwc)
wc.to_image()
```
