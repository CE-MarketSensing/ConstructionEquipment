```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service # 셀레니움 버전 4에서는 이걸 써야 한다. 
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd
from selenium.common.exceptions    import NoSuchWindowException
from selenium.webdriver.support.ui import WebDriverWait
```

```python
def chrome_setting():
    chrome_options = webdriver.ChromeOptions()
#     chrome_options.add_argument('--headless') # 크롬 탭이 나타나지 않게 옵션 조정
    chrome_options.add_argument('--no-sandbox')
    # 공유 메모리를 담당하는 디렉토리 사용 x
    chrome_options.add_argument('--disable-dev-shm-usage')
    # 본인 user-agent값 입력
    chrome_options.add_argument({"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36"})
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
    return driver
```

```python
options = Options()
options.add_experimental_option("detach", True) # 브라우저 바로 닫힘 방지
options.add_experimental_option('excludeSwitches', ['enable-logging']) #불필요한 메세지 제거
```

### 1-1 기사 링크 먼저 모으기

```python
url = "https://www.liebherr.com/en/usa/latest-news/news-press-releases/news-press-releases.html?page=1"

# url 로 이동
driver.get(url)
time.sleep(3)
```

```python
html = driver.page_source
soup = BeautifulSoup(html, "lxml")

articles = soup.find_all("section", attrs={"class":"news_teaser_module"})
```

```python
# //*[@id="pageWrapper"]/div[2]/div/form/article[2]/div/section[2]/div/div[2]/p[2]

for article in articles:
    print(article.find("p", attrs={"class":None}).text)
```

```python
title  = []
date  = []
sumerize = []
link = []

for article in articles:
    print("TITLE : ", article.a.get_text())
    # print("DATE : ", article.h2.previous_sibling.previous_sibling.get_text())
    print("sumerize : ", article.h2.find_next_sibling("p").text)
    print("Link : ", "https://www.liebherr.com"+article.a.get('href'))

    title.append(article.a.get_text())
    date.append(article.h2.previous_sibling.previous_sibling.get_text())
    sumerize.append(article.h2.find_next_sibling("p").text)
    link.append("https://www.liebherr.com"+article.a.get('href'))
    time.sleep(1)
    
news = pd.DataFrame({"제목": title, "날짜" : date, "링크":link, "요약":sumerize})
print(news)
```

```python
title  = []
date  = []
summarize = []
link = []

for page in range(1, 83):

    # url 로 이동
    url = f"https://www.liebherr.com/en/usa/latest-news/news-press-releases/news-press-releases.html?page={page}"
    driver.get(url)
    time.sleep(5)

    # html 가져오기
    html = driver.page_source
    soup = BeautifulSoup(html, "lxml")

    # 기사내용 10개씩
    articles = soup.find_all("section", attrs={"class":"news_teaser_module"})

    for article in articles:
        title.append(article.a.get_text())
        date.append(article.h2.previous_sibling.previous_sibling.get_text())
        summarize.append(article.h2.find_next_sibling("p").text)
        link.append("https://www.liebherr.com"+article.a.get('href'))
        time.sleep(1)

news = pd.DataFrame({"제목": title, "날짜" : date, "링크":link, "요약":summarize})
print(news)
```

### 1-2. 기사 링크로 들어가서 내용 스크래핑

```python
url = "https://www.liebherr.com/en/usa/latest-news/news-press-releases/detail/successful-participation-in-mro-americas-2023.html"
driver.get(url)

# driver.find_element(By.CSS_SELECTOR, "[aria-label='Decline']").click()
# time.sleep(1)

html = driver.page_source
soup = BeautifulSoup(html, "lxml")
print("title : ", soup.h1.text)

top_page = soup.select(".top_of_page")


top_sum = top_page[0].select(".bold")[0].find_all("li")

for p in top_sum:
    print(p.text)

for p in top_page[0].select("p.bold"):
    print(p.text)

news2 = soup.select("#news2 > div")
for n in news2:
    print(n.text)
```

```python
summerize = []
heads = []
contents = []

for link in news["링크"]:
    sums = []
    bolds = []
    content = []

    driver.get(link)
    time.sleep(1)

    html = driver.page_source
    soup = BeautifulSoup(html, "lxml")
    top_page = soup.select(".top_of_page")

    try:
        for p in top_page[0].select(".bold")[0].find_all("li"):
            sums.append(p.text)

        for p in top_page[0].select("p.bold"):
            bolds.append(p.text)
    except:
        print(">>> 예외발생")
        print(link)


    for p in soup.select("#news2 > div"):
        content.append(p.text)


    if len(sums) > 0 :
        summerize.append(". ".join(sums))

    if len(bolds) > 0 :
        heads.append(bolds[0])

    contents.append(". ".join(content))


df = pd.DataFrame({"토픽": summerize, "중요" : heads, "내용":contents})
df.head()
```
